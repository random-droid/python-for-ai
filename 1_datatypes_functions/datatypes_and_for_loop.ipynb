### Lists and for loop

A **list** is an ordered collection of items. Think of it as a row of numbered boxes — each box holds a value, and each box has an *index* (its position number).

`revenues = [50, 60, 55]`

```
 index:     0      1      2          <- position (always starts at 0!)
          +------+------+------+
 value:   |  50  |  60  |  55  |      <- revenues
          +------+------+------+
 from end:   -3     -2     -1         <- negative index counts backwards
```

So:
- `revenues[0]` -> `50`  (the **first** item, not the "1st"!)
- `revenues[2]` -> `55`  (the third item)
- `revenues[-1]` -> `55` (the **last** item — handy shortcut)
- `len(revenues)` -> `3` (how many boxes there are)

> ⚠️ The #1 beginner gotcha: counting starts at **0**, not 1. A list of 3 items has indexes 0, 1, 2 — there is no index 3. (Ask for `revenues[3]` and Python throws an `IndexError`.)

````python
revenues = [50, 60, 55]
expenses = [10, 30, 35]
````

````python
type(revenues)
````

````python
len(revenues)
````

````python
for rev in revenues:
    print(rev)
````

````python
for i in range(len(revenues)):
    profit = revenues[i] - expenses[i]
    margin = profit * 100 / revenues[i]
    print(margin)
````

### Dict and Tuples

````python
fin = {"revenue": 50, "expenses": 10}

profit = fin["revenue"] - fin["expenses"]
margin = profit * 100 / fin["revenue"]
margin
````

````python
financials = [
    {"revenue": 50, "expenses": 10},
    {"revenue": 60, "expenses": 30},
    {"revenue": 55, "expenses": 35}
]

for fin in financials:
    profit = fin["revenue"] - fin["expenses"]
    fin['margin'] = profit * 100 / fin["revenue"]
````

````python
financials
````



**Nested Dictionary**

````python
financials = {
    "Q1": {"revenue": 50, "expense": 10},
    "Q2": {"revenue": 60, "expense": 30},
    "Q3": {"revenue": 55, "expense": 35},
}

for quarter, data in financials.items():
    margin = (data["revenue"] - data["expense"]) / data["revenue"] * 100
    print(f"{quarter}: {margin:.2f}%")
````

### If condition

````python
financials = {
    "Q1": {"revenue": 50, "expense": 10},
    "Q2": {"revenue": 60, "expense": 30},
    "Q3": {"revenue": 55, "expense": 35},
    "Q4": {"revenue": -10, "expense": 35},
}

for quarter, data in financials.items():
    if data["revenue"]<0:
        print(f"Invalid revenue for {quarter}. Skipping.")
        continue
    margin = (data["revenue"] - data["expense"]) / data["revenue"] * 100
    print(f"{quarter}: {margin:.2f}%")
````
