# Calling an LLM (Google Gemini)

### First, a 2-minute idea: objects & methods

In a moment you'll write `client = genai.Client()` and then `client.models.generate_content(...)`. Before that looks like alien syntax, here's the only mental model you need:

- A **class** is a *blueprint* — like the design of a car. 🚗
- An **object** is one actual thing built from that blueprint. You build it by calling the blueprint with `()`:  `my_car = Car()`.
- A **method** is something that object can *do*. You call it with a dot:  `my_car.start()`, `my_car.brake()`.

That's it. The dot (`.`) means *"reach into this object and use one of its abilities."*

So when we write:

```python
client = genai.Client()              # build a Gemini client object from the library's blueprint
response = client.models.generate_content(...)   # ask that object to do something for us
```

...we're just *using* an object the `google-genai` library gives us.

> 💡 You will **not** write your own classes today. Almost every AI library hands you ready-made objects like this — your job is just to create one with `()` and call its methods with `.`. (And when you eventually do need a custom class, you'll let AI write it for you.)

````python
from google import genai
from google.genai import types

client = genai.Client()
response = client.models.generate_content(
    model="gemini-3.5-flash",
    config=types.GenerateContentConfig(
        system_instruction="You are a helpful assistant who specializes in food history. answer as if you are a CEO say Elon Musk"
    ),
    contents="Tell me who invented Samosa in 3 lines"
)
print(response.text)
````

````python
client = genai.Client()

def generate_recipe(ingradients, cuisine, diet):
    '''
    Generate a recipe using ingradients provided. 
    Cuisine can be "Italian", "Mexican", "Indian", "Any" etc.
    diet can be "Vegan", "Vegeterian", "Jain", "Any" etc.
    '''
    
    prompt = f'''
    Generate one food receipe using these ingradients: {','.join(ingradients)}
    Receipe should not be more than 100 words
    Cuisine: {cuisine}
    Diet: {diet}
    '''
    response = client.models.generate_content(
        model="gemini-3.5-flash",
        contents=prompt
    )
    return response.text       
````

````python
recipe = generate_recipe(["Tomatoes", "lentils", "onion", "jeera", "salt"],"Indian","Vegan")
recipe
````
