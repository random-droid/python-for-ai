# Understanding Classes — Objects & Methods

## 1. Life without classes (the pain) 😩

Suppose you are building a system to manage employee information.

````python
# Employee 1
name1 = "Mohan"
salary1 = 80000

# Employee 2
name2 = "Fatima"
salary2 = 95000

def calculate_bonus(salary):
    return salary * 0.10

print(f"{name1}'s bonus: {calculate_bonus(salary1)}")
print(f"{name2}'s bonus: {calculate_bonus(salary2)}")
````

Now imagine you have 500 employees. You will end up with 
```
name1, salary1
name2, salary2
name3, salary3
...
name500, salary500
```
This becomes difficult to:

* Store employee data together
* Pass employee information around
* Add new properties such as department, age, email
* Add employee-related behaviors such as bonus calculation, promotion, etc.

The Pain Point

An employee has multiple pieces of information:

```
name
salary
department
email
age
```

and multiple actions:

```
calculate_bonus()
promote()
send_email()
```
Keeping all these variables and functions separate becomes messy.

## 2. The fix: a class bundles data + behavior

A class allows us to bundle data and behavior together.

````python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def calculate_bonus(self):
        return self.salary * 0.10
````

````python
emp1 = Employee("John", 80000)
emp2 = Employee("Mary", 95000)

print(emp1.calculate_bonus())
print(emp2.calculate_bonus())
````

## 4. A second style of class: data models (Pydantic) 📋

The `Phone` class bundled **data + behavior** — it could `call()` and `charge()`. But often you just want a class that holds **data in a guaranteed shape**, with no behavior at all. That's where **Pydantic** comes in, and you'll see it *everywhere* in AI code.

Remember type hints from the functions lesson (`revenue: float`)? Back then Python *ignored* them. A Pydantic model uses those same labels — but **enforces** them.

````python
from pydantic import BaseModel

# A different style of class: just LIST the fields and the type each must hold.
# No __init__, no self, no methods — Pydantic writes all that for you.
class Recipe(BaseModel):
    title: str
    calories: int

# Build one like any other object:
dal = Recipe(title="Yellow Dal", calories=280)
print(dal.title)
print(dal.calories)
````

Two new bits of syntax:
- `class Recipe(BaseModel)` — the `(BaseModel)` means *"borrow Pydantic's superpowers."* (This is **inheritance**: your class gets all of `BaseModel`'s abilities for free.)
- `title: str` — each field is labeled with the type it must hold, exactly like the function hints you just learned.

````python
# The payoff: Pydantic ENFORCES the types. Hand it junk and it refuses:
Recipe(title="Dal", calories="not a number")   # 💥 ValidationError
````

💥 `ValidationError`! Unlike plain function hints, Pydantic actually checks the data and rejects anything that doesn't fit the shape.

That guarantee is gold when data comes from an *unpredictable* source... like an **LLM**. In the next lessons we'll ask an AI to fill in a Pydantic model, and Pydantic guarantees we get back exactly the shape we asked for. You'll also meet it again when we build a **FastAPI** backend. So this little class is a big deal. 🚀
