### Type hints (labeling your inputs and outputs)

Look at our function: `def calculate_margin(revenue, expenses)`. Nothing tells you — or the next developer — what `revenue` is meant to be. A number? A list? Text?

Python lets you add optional **type hints**: tiny labels saying what type you expect.

```python
def calculate_margin(revenue: float, expenses: float) -> float:
    profit = revenue - expenses
    return profit * 100 / revenue

calculate_margin(50, 10)
```

Two new bits:
- `revenue: float` — *"this input should be a float"*
- `-> float` — *"this function returns a float"*

> ⚠️ **The catch:** type hints are just labels for humans (and tools). Python does **not** enforce them. Watch — we promise an `int`, hand it a string, and Python happily runs anyway:

```python
def double(x: int) -> int:
    return x * 2

print(double("ha"))   # we said int... gave a string... Python shrugs -> 'haha'
```

So why bother? Because hints make your code **self-documenting**, your editor autocompletes better, and AI coding assistants write *far* better code when they can see the types.

😏 Foreshadow: in the **classes** lesson you'll meet **Pydantic**, where these very same labels suddenly *do* get enforced.

### args and kargs

```python
def calc_total_expense(exp1,exp2):
    return exp1+exp2

calc_total_expense(2,3)    
```

```python
def calc_total_expense(*args):
    print(args)
    return sum(args)

calc_total_expense(2,3,5)    
```

```python
def total_expenses(rent, phone_bill):
    return rent + phone_bill

total_expenses(rent=1000, phone_bill=30)    
```

```python
def total_expenses(**kargs):
    total = 0
    for key, value in kargs.items():
        total += value
    return total

total_expenses(rent=1000, phone_bill=30, food=60)
```
